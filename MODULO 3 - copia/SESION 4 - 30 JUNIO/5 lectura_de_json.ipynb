### 2. Desarrollo:

```python
import pandas as pd
```

Para leer archivos en formato JSON se usa la forma:

`pd.read_json(-ruta-)`

La ruta (o `path`) puede ser absoluto o relativo (en el Prework hay un link donde se explica esto con más claridad), ahora lee los datos del archivo `zomato_reviews-clean.json` que se encuentra en la carpeta de `Datasets`.

```python
df = pd.read_json("zomato_reviews-clean.json")

df
```

Ahora asignamos nuevos índices para contar a partir de 1, para ello se usa la función `df.set_index(-nuevo índice como serie-)` y para generar la serie se puede usar `range()`, sólo que hay qye tener cuidados de los límites, veamos como:

```python
df_1 = df.set_index (pd.Series (range(1, 1181)))   # 1,......1181

df_1
```

Investigamos el nombre de las columnas con la propiedad `df.columns` del dataframe `df_1`:

```python
df.columns
```

O eliminamos la columna `price_range` con `dataframe.drop(columns=-lista-)`:

```python
df_2 = df_1.drop(columns=["price_range"])

df_2
```
