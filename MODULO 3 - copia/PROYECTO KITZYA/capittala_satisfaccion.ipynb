```python
# se eliminan las columnas con valores float que tienen demasiados NaN's y que para este analisis no servirán
# Se eliminan columnas duplicadas

df_1 = df.drop(columns=["tranquilidad", "plusvalia", "distribucion-espacios", "ubicacion", "precio-promocion", "calidad", "decision_compra1", "razon_busqueda1"])
df_1
```

```python
# La columna fecha está en tipo object, cambiarlo a tipo fecha

diccionario_de_conversion = {
    "fecha":"datetime64[ns]"
}
```

```python
df_1 = df_1.astype(diccionario_de_conversion)

df_1.head()
```

```python
# Comprobamos que se haya cambiado el tipo de valor de fecha
df_1.dtypes
```

```python
# En las columnas de algo_que_cambiar encontramos que habías varios NaN's por lo que damos por hecho que las personas encuestadas no quieren vambiar nada y colocaremos "nada" esos campos

df_1['algo_que_cambiar'] = df_1['algo_que_cambiar'].fillna("nada")
df_1.head()
```

```python
# De la misma forma, en la columna proyectos_visitados_antes_de_compra encontramos varios datos vacios por lo que damos por hecho que los clientes no lo llenaron porque no habían
# visitado ningún otro proyecto, así que llenamos los espacios con "ninguno"

df_1['proyectos_visitados_antes_de_compra'] = df_1['proyectos_visitados_antes_de_compra'].fillna("ninguno")
df_1.head()
```
