```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df = pd.read_csv('../../Datasets/diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
ax1 = sns.displot(df['glucose'], kde=False)
ax1.set(title='Concentración de Glucosa')
ax1.set(xlabel='cantidad de glucosa', ylabel='total personas');
```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
ax2 = sns.barplot(x=pregnancy_counts.index, y=pregnancy_counts)

ax2.set(title='Número de veces Embarazada')
ax2.set(xlabel='Cantidad de Embarazos', ylabel='Cantidad de Personas');
```

### c)

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
sns.kdeplot(healthy['glucose'], shade=True)
sns.kdeplot(sick['glucose'], shade=True)

healthy.set_title('Sanos vs Enfermos', fontsize=13, pad=15);
sns.set(xlabel='número de personas');
sns.legend(loc='upper right');
```

### d)

```python
sns.kdeplot(sick['blood_pressure'], shade=True)
sns.kdeplot(sick['glucose'], shade=True)
sns.kdeplot(sick['insulin'], shade=True)
```
